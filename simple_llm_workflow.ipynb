```python
import os
os.environ["LANGGRAPH_DISABLE_DAG_RENDER"] = "1"
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from google import genai
```

```python
load_dotenv()
```

```python
import os
api_key = os.getenv("GEMINI_API_KEY_SELF") # or os.getenv("GEMINI_API_KEY")
if not api_key:
	raise EnvironmentError("GEMINI_API_KEY2 (or GEMINI_API_KEY) not set in environment. Add it to your .env or export it.")

# client = genai.Client(api_key=api_key)
```

```python
model = ChatOpenAI(
   model="gemini-2.5-flash",
   api_key=api_key,
   base_url="https://generativelanguage.googleapis.com/v1beta/openai"
)
```

```python
#state
class LLMState(TypedDict):
    que: str
    ans: str
```

```python
def llmqa(state: LLMState)-> LLMState:
    #extract que from state
    que= state['que']
    #form a prompt
    prompt= f'answer the following question {que}'

    ans= model.invoke(prompt).content
    state['ans']= ans
    return state
```

```python
#create graph
graph= StateGraph(LLMState)

graph.add_node('llm_que', llmqa)

graph.add_edge(START, 'llm_que')
graph.add_edge('llm_que', END)

workflow= graph.compile()
```

```python
inital_state= {'que': 'how far is sun from earth?'}

final_state= workflow.invoke(inital_state)
print(final_state['ans'])
```
